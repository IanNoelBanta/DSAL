```python
# import heapq module
from heapq import *

setOfNumbers = [55, 25, 45, 75, 5, 65]
print("Set of numbers before being converted into a heap. ", setOfNumbers)
# Use heapify to rearrange the elements
heapify(setOfNumbers)
print("A heap data structure ", setOfNumbers)

# Insert new node
heappush(setOfNumbers, 35)
print("A heap with a newly inserted value ", setOfNumbers)
heapify(setOfNumbers)
print("A rotated heap data structure", setOfNumbers)

# Delete root node
heappop(setOfNumbers)
print("A heap after deletion of root node ", setOfNumbers)
heapify(setOfNumbers)
print("A rotated heap data structure", setOfNumbers)
```
