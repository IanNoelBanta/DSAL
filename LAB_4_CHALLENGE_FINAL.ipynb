```python
import array as arr
from collections import deque ## for stack, queue, and deque


class array:
    def __init__(self):
        self.iAmAnArray = list()
        print('Container Array is now available.')

    def getValue(self):
        index = int(input('On what index do you want to insert a value? '))
        value = input('Please insert a value to the array: ')
        self.iAmAnArray.insert(index, value) ## inserts a value in the corresponding index

    def displayArray(self):
        if not self.iAmAnArray:
            print('Container Array is empty.')
        else:
            print(f'Container Array contains: {self.iAmAnArray}')

    def displayByIndex(self):
        index = 0
        for element in self.iAmAnArray: ## prints out the elements by their index
            print(f'Index {index} in the array is: {element}')
            index += 1

    def removeIndex(self):
        if not self.iAmAnArray:
            print('Container Array is empty.')
        else:
            index = int(input('What index would you like to remove? '))
            print(f'You removed {self.iAmAnArray.pop(index)} from the array') ## removes a value from the given index
            print(f'Updated: {self.iAmAnArray}')


class stack:
    def __init__(self):
        self.iAmAStack = deque()
        print('Container Stack is now available.')

    def displayStack(self):
        print(f'Container Stack contains: {self.iAmAStack}')

    def push(self):
        value = input('Please insert a value on top of the stack: ')
        self.iAmAStack.append(value) ## inserts a value on top of the stack

    def pop(self):
        print(f'You popped {self.iAmAStack.pop()} from the top of the stack.') ## prints out the popped value from the top of the stack

    def peep(self):
        print(f'The element on top of the stack is {self.iAmAStack[-1]}') ## prints the last/top value


class queue:
    def __init__(self):
        self.iAmAQueue = deque()
        print('Container Queue is now available.')

    def displayQueue(self):
        print(f'Container Queue contains: {self.iAmAQueue}')

    def enqueue(self):
        value = input('Please insert a value to the tail of the queue: ')
        self.iAmAQueue.appendleft(value) ## inserts a value to the left side
        print(f'Updated Queue: {self.iAmAQueue}')

    def dequeue(self):
        print(f'You popped {self.iAmAQueue.pop()} from the head of the queue.') ## prints out the popped value from the right side
        print(f'Updated Queue: {self.iAmAQueue}')


class dq:
    def __init__(self):
        self.iAmAnDeque = deque()
        print('Container Stack is now available.')

    def displayDeque(self):
        print(f'Container Deque contains: {self.iAmAnDeque}')

    def insertleft(self):
        value = input("Please insert a value to the left end of the deque: ")
        self.iAmAnDeque.appendleft(value) ## inserts a value to the left side

    def insertright(self):
        value = input("Please insert a value to the right end of the deque: ")
        self.iAmAnDeque.append(value) ## inserts a value to the right side

    def popleft(self):
        print(f'You popped {self.iAmAnDeque.popleft()} from the left end of the deque.') ## pops a value from the left side

    def popright(self):
        print(f'You popped {self.iAmAnDeque.pop()} from the right end of the deque.') ## pops a value from the right side


## instantiate objects for each class/container
containerArray = array()
containerStack = stack()
containerQueue = queue()
containerDeque = dq()


def arraywattodo():
    print('''What do you want to do with the Array?''')
    stay = True
    while stay:
        print('''
        Press 1 - Insert a value to the array
        Press 2 - Display the elements in the array
        Press 3 - Display the elements by their index
        Press 4 - Remove an element by its index
        ''')

        move = int(input('Please pick: '))
        if move == 1:
            containerArray.getValue()
            containerArray.displayArray()
            containerStack.displayStack()
            containerQueue.displayQueue()
            containerDeque.displayDeque()
        elif move == 2:
            containerArray.displayArray()
        elif move == 3:
            containerArray.displayByIndex()
        else:
            containerArray.removeIndex()

        exit = input('Do you want try other functions of the array? [y/n] ')
        if exit.lower() == 'n':
            stay = False


def stackwattodo():
    print('''What do you want to do with the Stack?''')
    stay = True
    while stay:
        print('''
        Press 1 - Push a value on top of the stack
        Press 2 - Pop the value on top of stack
        Press 3 - Peep the element on top of the stack
        Press 4 - Display the elements in the stack
        ''')

        move = int(input('Please pick: '))
        if move == 1:
            containerStack.push()
            containerArray.displayArray()
            containerStack.displayStack()
            containerQueue.displayQueue()
            containerDeque.displayDeque()
        elif move == 2:
            containerStack.pop()
        elif move == 3:
            containerStack.peep()
        else:
            containerStack.displayStack()

        exit = input('Do you want try other functions of the stack? [y/n] ')
        if exit.lower() == 'n':
            stay = False


def queuewattodo():
    print('''What do you want to do with the Queue?''')
    stay = True
    while stay:
        print('''
        Press 1 - Insert a value to the tail of the queue
        Press 2 - Pop the value from the head of the queue
        Press 3 - Display the elements in the queue
        ''')

        move = int(input('Please pick: '))
        if move == 1:
            containerQueue.enqueue()
            containerArray.displayArray()
            containerStack.displayStack()
            containerQueue.displayQueue()
            containerDeque.displayDeque()
        elif move == 2:
            containerQueue.dequeue()
        else:
            containerQueue.displayQueue()

        exit = input('Do you want try other functions of the queue? [y/n] ')
        if exit.lower() == 'n':
            stay = False


def dequewattodo():
    print('''What do you want to do with the deque?''')
    stay = True
    while stay:
        print('''
        Press 1 - Insert a value to the left end of the deque
        Press 2 - Insert a value to the right end of the deque
        Press 3 - Pop a value from the left end of the deque
        Press 4 - Pop a value from the right end of the deque
        Press 5 - Display the elements in the deque
        ''')

        move = int(input('Please pick: '))
        if move == 1:
            containerDeque.insertleft()
            containerArray.displayArray()
            containerStack.displayStack()
            containerQueue.displayQueue()
            containerDeque.displayDeque()
        elif move == 2:
            containerDeque.insertright()
            containerArray.displayArray()
            containerStack.displayStack()
            containerQueue.displayQueue()
            containerDeque.displayDeque()
        elif move == 3:
            containerDeque.popleft()
        elif move == 4:
            containerDeque.popright()
        else:
            containerDeque.displayDeque()

        exit = input('Do you want try other functions of the deque? [y/n] ')
        if exit.lower() == 'n':
            stay = False


def main():
    print('''\t  Available Containers
        Press 1 - Array
        Press 2- Stack
        Press 3 - Queue
        Press 4 - Deque
            ''')

    addmore = True
    while addmore:
        pick = int(input('Pick a container: '))
        if pick == 1:
            arraywattodo()
        elif pick == 2:
            stackwattodo()
        elif pick == 3:
            queuewattodo()
        else:
            dequewattodo()

        more = input('Do you want to add another value? [y/n] ')
        if more.lower() == 'n':
            addmore = False


main()
```
