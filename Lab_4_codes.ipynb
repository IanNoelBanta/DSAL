##Lab 4_4.py

```python
class Stack1:

    def __init__(self):
        self.stack = list()

    # Insert method to add element

    def insert(self, value):
        if value not in self.stack:
            # self.queue.insert(0, value)  # This is wrong.
            # self.stack.insert(len(self.stack), value)
            self.stack.append(value)
            return True
        return False

    def size(self):
        return len(self.stack)

    def remove_item(self):
        if len(self.stack) > 0:
            return self.stack.pop()
        return "No elements in Queue!"

    def display_entire_stack(self):
        return self.stack

    def update_value(self, index, value):
        self.stack.pop(index)
        self.stack.insert(index, value)


stack1 = Stack1()
for number in range(10):
    stack1.add_item(number)
    print(stack1.display_entire_stack())
for number in range(10):
    stack1.remove_item()
    print(stack1.display_entire_stack())
```

##Lab 4_7.py

```python
class Deque1:

    def __init__(self):
        self.deque = list()

    # Insert method to add element

    def add_item(self, value, end):
        if value not in self.deque:
            if end == 0:  # if end = 0. it's the front
                self.deque.insert(0, value)
            elif end == 1:  # if end = 0. it's the rear
                self.deque.insert(len(self.deque), value)
            return True
        return False

    def size(self):
        return len(self.deque)

    def remove_item(self):
        if len(self.deque) > 0:
            return self.deque.pop()
        return "No elements in Queue!"

    def display_entire_stack(self):
        return self.deque

    def update_value(self, index, value):
        self.deque.pop(index)
        self.deque.insert(index, value)


deque1 = Deque1()
for number in range(10):
    deque1.add_item(number, 1)
    print(deque1.display_entire_stack())
```
