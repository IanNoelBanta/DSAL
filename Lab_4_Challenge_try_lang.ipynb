Create a new python file and name it Lab4_Challenge.py.
Create a program that can allow a user to insert one value to any of the 4 containers namely iAmAnArray, iAmAStack, iAmAQueue, iAmAnDeque.
The containers stated above should be based on the behavior of the ADT which they are named after.

Since the containers above should behave as their ADT counterpart, some containers may ask on what index or position will the value be inserted.
After each successful insertion, the contents of all the 4 containers must be displayed.

```python
class Array:

    def __init__(self):
      self.container = []
      print('Container Array is now available.')
    
    def getValue(self):
      value = input('Please insert a value to the array: ')
      self.container.append(value)
    
    def displayArray(self):
      print(f'Container Array contains: {self.container}')

class Stack:

    def __init__(self):
      self.container = []
      print('Container Stack is now available.')
    
    def getValue(self):
      value = input('Please push a value to the Stack: ')
      self.container.append(value)
  
    def displayStack(self):
      print(f'Container Stack contains: {self.container}')
    
    def size(self):
      print(len(self.container))
    
    def remove(self):
      return self.container.pop()





class Queue:

    def __init__(self):
      value = []
      print('Container Queue is now available.')

class Deque:

    def __init__(self):
      value = []
      print('Container Deque is now available.\n')


objectArray = Array()
objectStack = Stack()
objectQueue = Queue()
objectDeque = Deque()

addMore = True

print('Where do you want to insert a value? ')
while addMore:

  pick = int(input('''\tPress 1 for Array 
\tPress 2 for Stack
\tPress 3 for Queue
\tPress 4 for Deque
\n'''))
  
  if pick == 1:
    objectArray.getValue()
    objectArray.displayArray()
    objectStack.displayStack()
  elif pick == 2:
    objectStack.getValue()
    objectStack.displayStack()
    asktoremove = input('[y/n]')
    if asktoremove.lower() == 'y':
      objectStack.remove()
      print(f'Your updated Stack: {objectStack.displayStack()}')

  elif pick == 3:
    objectQueue.getValue()
  else:
    objectDeque.getValue()
  
  more = input('Do you want to insert another value? [y/n]')
  if more.lower() == 'n':
    addMore = False

    

```
