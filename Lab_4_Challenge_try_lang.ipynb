Create a new python file and name it Lab4_Challenge.py.
Create a program that can allow a user to insert one value to any of the 4 containers namely iAmAnArray, iAmAStack, iAmAQueue, iAmAnDeque.
The containers stated above should be based on the behavior of the ADT which they are named after.

Since the containers above should behave as their ADT counterpart, some containers may ask on what index or position will the value be inserted.
After each successful insertion, the contents of all the 4 containers must be displayed.

```python
##Class for Array
class Array:

    def __init__(self):
      self.container = []
      print('Container Array is now available.')
    
    def getValue(self):
      value = input('Please insert a value to the array: ')
      self.container.append(value)
    
    def displayArray(self):
      if not self.container:
        print("Container Array is empty.")
      else:
        print(f'Container Array contains: {self.container}')
    
    def displayByIndex(self):
      index = 0
      for value in self.container:
        print(f'Index {index} in the array is: {value}')
        index += 1
    
    def removeIndex(self):
      if len(self.container) > 0:
        index = int(input('What index would you like to remove?' ))
        self.container.pop(index)
        print(f'Here is your updated array: {self.container}')
      else:
        print("The container is empty")


##Class for Stack
class Stack:

    def __init__(self):
      self.container = []
      print('Container Stack is now available.')
    
    def getValue(self):
      value = input('Please push a value to the Stack: ')
      self.container.append(value)
  
    def displayStack(self):
      if not self.container:
        print("Container Stack is empty.")
      else:
        print(f'Container Stack contains: {self.container}')
    
    def size(self):
      print(f'Size: {len(self.container)}')
    
    def remove(self):
      last = self.container.pop()
      print(f'{last} is popped from the stack.')
      print(f'Updated stack {self.container}')


##Class for Queue
class Queue:

    def __init__(self):
      self.container = []
      print('Container Queue is now available.')

    def getValue(self):
      value = input('Please insert a value to the Queue: ')
      if value not in self.container:
        self.container.append(value)
      else:
        print(f'{value} is already in the queue.')
    
    def displayQueue(self):
      print(f'Container Queue contains: {self.container}')

    def size(self):
      print(f'Size: {len(self.container)}')
    
    def remove(self):
      if len(self.container) > 0:
        last = self.container.pop(0)
        print(f'{last} is popped from the queue.')
        print(f'Updated queue {self.container}')
      else:
        print('The queue is empty.')


##Class for Deque
class Deque:

    def __init__(self):
      self.container = []
      print('Container Deque is now available.\n')

    def append(self):
      right = input('Please push a value to the Deque: ')
      self.container.append(-1, right)

    def appendLeft(self):
      left = input('Please push a value to the Deque: ')
      self.container.append(0, left)

    def remove(self):
      self.container.pop(-1)
    
    def removeLeft(self):
      self.container.pop(0)

    def displayStack(self):
      print(f'Container Deque contains: {self.container}')
    
    def size(self):
      print(f'Size: {len(self.container)}')


##instantiate an object for all classes
objectArray = Array()
objectStack = Stack()
objectQueue = Queue()
objectDeque = Deque()


##Ask the user
print('Where do you want to insert a value? ')
print('''\tPress 1 for Array 
\tPress 2 for Stack
\tPress 3 for Queue
\tPress 4 for Deque
''')


##variable for the loop
addMore = True

##While loop - stop if variable is false
while addMore:
  ##ask user 
  pick = int(input('\nPlease pick a container: '))
  
  if pick == 1: ##inserts in array
    objectArray.displayArray()
    objectArray.getValue()
    objectArray.displayArray()
    objectStack.displayStack()
    asktoremove = input('Would you like to remove a value in the array? ')
    if asktoremove.lower() == 'y':
      objectArray.removeIndex()

    asktoindex = input('Do you want to display all the values in the array by their index? [y/n] ')
    if asktoindex.lower() == 'y':
      objectArray.displayByIndex()


  elif pick == 2: ##inserts in stack
    objectStack.displayStack()
    print(f'Size of the stack: {objectStack.size()}')
    objectStack.getValue()
    objectArray.displayArray()
    objectStack.displayStack()
    asktoremove = input('Do you want to remove the last value in the stack [y/n]')
    if asktoremove.lower() == 'y':
      objectStack.remove()
    asksize = input('Do you want to know the current size of the stack? [y/n]')
    if asksize.lower() == 'y':
      objectStack.size()

  elif pick == 3: ##inserts in queue
    objectQueue.displayQueue
    objectQueue.getValue()

  else: ##inserts in deque
    objectDeque.getValue()
  
  ##insert more??
  more = input('\nDo you want to insert another value? [y/n] ')
  if more.lower() == 'n':
    addMore = False ##change to false to exit loop 

    

```
